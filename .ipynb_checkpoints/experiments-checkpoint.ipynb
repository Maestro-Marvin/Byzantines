```python
import numpy as np
import scipy.stats as sps
import random
from tqdm import trange
import matplotlib.pyplot as plt
from tqdm import tqdm
from sklearn.metrics import accuracy_score

seed = 42
rng = np.random.default_rng(seed)
```

```python
dataset = "mushrooms.txt"

from sklearn.datasets import load_svmlight_file
data = load_svmlight_file(dataset)
X, y = data[0].toarray(), data[1]

y = y - 1
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=seed)
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def cross_entropy_loss(w, X, Y):
    return -Y * np.log(sigmoid(X @ w)) - (1 - Y) * np.log(1 - sigmoid(X @ w))

def f(w, X, Y, lambd):
    return np.mean(cross_entropy_loss(w, X, Y)) + (lambd / 2) * np.linalg.norm(w) ** 2

def grad(w, X, Y, lambd):
    return X.T @ (sigmoid(X @ w) - Y) / X.shape[0] + lambd * w

def prediction(w, X):
    return sigmoid(X @ w) > 1/2
```

```python
class BaseAggregator(object):
    def __init__(self):
        pass
    def __call__(self, inputs):
        pass
    def __str__(self):
        pass

class MeanAggregator(BaseAggregator):
    def __call__(self, inputs):
        return np.mean(inputs, axis=0)
    def __str__(self):
        return "mean agg"

class GoodMeanAggregator(BaseAggregator):
    def __init__(self, devices):
        devices_types = [isinstance(device, GoodWorker) for device in devices]
        self.devices_types = devices_types
    def __call__(self, inputs):
        return np.mean(inputs[self.devices_types], axis=0)
    def __str__(self):
        return "good mean agg"
        
class MedianAggregator(BaseAggregator):
    def __call__(self, inputs):
        return np.array([np.median([inputs[:, i]]) for i in range(inputs.shape[1])])
    def __str__(self):
        return "median agg" 

class TrimmedMean(BaseAggregator):
    def __init__(self, delta=0.1):
        self.delta = delta
    def __call__(self, inputs):
        lb = int(self.delta * len(inputs))
        ub = len(inputs) - lb
        return np.array([np.mean(np.sort(inputs[:, i])[lb : ub]) for i in range(inputs.shape[1])])
    def __str__(self):
        return f"trimmed mean delta={self.delta}" 
        
class CCAggregator(BaseAggregator):
    def __init__(self, clip_range=1, num_iter=10):
        self.clip_range = clip_range
        self.num_iter = num_iter
    def __call__(self, inputs):
        v = np.mean(inputs, axis=0)
        for i in range(self.num_iter):
            c = []
            for m in inputs:
                c.append((m - v) * min(1, self.clip_range / np.linalg.norm(m - v)))
            c = np.array(c)
            v += np.mean(c, axis=0)
        return v
    def __str__(self):
        return f"cc agg r={self.clip_range}, l={self.num_iter}" 
```

```python
class BaseWorker(object):
    def __init__(self):
        pass
    def __call__(self,  w, X_batch, y_batch, lambd):
        pass
    def __str__(self):
        pass

class GoodWorker(BaseWorker):
    def __init__(self, beta):
        self.beta = beta
        self.m = 0
    def __call__(self, w, X_batch, y_batch, lambd):
        g = grad(w, X_batch, y_batch, lambd)
        self.m = (1 - self.beta) * g + self.beta * self.m
        return self.m
    def __str__(self):
        return "no attack"

class BadWorker(BaseWorker):
    def __init__(self):
        pass
    def __call__(self,  w, X_batch, y_batch, lambd):
        pass
    def __str__(self):
        pass

class RandomNoiseAttack(BadWorker):
    def __init__(self, noise_level):
        self.noise_level = noise_level
    def __call__(self, w, X_batch, y_batch, lambd):
        g = grad(w, X_batch, y_batch, lambd)
        noise = np.random.normal(0, self.noise_level, size=g.shape)
        return g + noise
    def __str__(self):
        return f"RN attack (noise={self.noise_level})"

class IPMAttack(BadWorker):
    def __init__(self, epsilon, good_gradients=None):
        self.epsilon = epsilon
        self.good_gradients = good_gradients
    def set_good_gradients(self, good_gradients):
        self.good_gradients = good_gradients
    def __call__(self, w, X_batch, y_batch, lambd):
        return -self.epsilon * np.mean(self.good_gradients, axis=0)
    def __str__(self):
        return f"IMP attack (epsilon={self.epsilon})"

class ALittleIsEnoughAttack(BadWorker):
    """
    Args:
        n (int): Total number of workers
        m (int): Number of Byzantine workers
    """
    def __init__(self, n, m, good_gradients=None):
        self.good_gradients = good_gradients
        s = np.floor(n / 2 + 1) - m
        cdf_value = (n - m - s) / (n - m)
        self.z_max = sps.norm.ppf(cdf_value)
    def set_good_gradients(self, good_gradients):
        self.good_gradients = good_gradients
    def __call__(self, w, X_batch, y_batch, lambd):
        mu = np.mean(self.good_gradients, axis=0)
        std = np.std(self.good_gradients, axis=0)
        return mu - std * self.z_max
    def __str__(self):
        return "ALIE attack"
```

```python
def disbalanced_batch(prob_class_1, batch_size):
    # Выбираем индексы с учётом вероятности
    class_1_indices = np.where(y_train == 1)[0]
    class_0_indices = np.where(y_train == 0)[0]

    num_class_1 = int(batch_size * prob_class_1)
    num_class_0 = batch_size - num_class_1

    sampled_class_1 = np.random.choice(class_1_indices, num_class_1, replace=False)
    sampled_class_0 = np.random.choice(class_0_indices, num_class_0, replace=False)
    sampled_indices = np.concatenate([sampled_class_1, sampled_class_0])
    np.random.shuffle(sampled_indices)

    X_batch = X_train[sampled_indices]
    y_batch = y_train[sampled_indices]

    return X_batch, y_batch


def distributed_sgd(aggregator, devices, good_devices, prob_class_1_per_device=None,
                    learning_rate = 0.01, max_iter = 10000, batch_size = 32, lambd = 0.01, eps=1e-8):
    
    criterion = lambda x: np.linalg.norm(grad(x, X_train, y_train, lambd)) 

    # Инициализация весов
    w = np.zeros(X_train.shape[1])

    errors = []
    accuracy = []
    err_x_0 = criterion(w)
    errors.append(criterion(w) / err_x_0)
    
    for iteration in tqdm(range(max_iter)):
        gradients = []
        
        for idx, device in enumerate(devices):
            if prob_class_1_per_device is not None:
                prob_class_1 = prob_class_1_per_device[idx]
                X_batch, y_batch = disbalanced_batch(prob_class_1, batch_size)
            else:
                indices = np.random.choice(X_train.shape[0], batch_size, replace=False)
                X_batch = X_train[indices]
                y_batch = y_train[indices]
            
            # Вычисляем градиент
            if isinstance(device, (IPMAttack, ALittleIsEnoughAttack)):
                device.set_good_gradients(gradients[:good_devices])
            grad_device = device(w, X_batch, y_batch, lambd)
            gradients.append(grad_device)
        
        # Агрегация градиентов
        gradients = np.array(gradients)
        aggregated_gradient = aggregator(gradients)
        
        # Шаг SGD
        w -= learning_rate * aggregated_gradient
    
        errors.append(criterion(w) / err_x_0)
        accuracy.append(accuracy_score(y_test, prediction(w, X_test)))
        if errors[-1] < eps:
            break

    return w, errors, accuracy
```

```python
def plot_accuracy(aggregators, accuracy, beta, attack):
    plt.figure(figsize=(10, 6))
    for agg, acc in zip(aggregators, accuracy):
        plt.plot(range(len(acc)), acc, label=str(agg))
    plt.xlabel('Номер итерации')
    plt.ylabel('Accuracy')
    plt.title(r'$\beta={}$, {}'.format(beta, attack))
    plt.legend()
    plt.grid(True)
    plt.show()

def plot_criterion(aggregators, errors, beta, attack):
    plt.figure(figsize=(10, 6))
    for agg, error in zip(aggregators, errors):
        plt.plot(range(len(error)), error, label=str(agg))
    plt.yscale('log')
    plt.xlabel('Номер итерации')
    plt.ylabel('Критерий')
    plt.title(r'Сходимость SGD для разного типа агрегации, $\beta={}$, {}'.format(beta, attack))
    plt.legend()
    plt.grid(True)
    plt.show()

def run_experiment(num_devices, g, beta, bad_worker, plot, prob_class_1_per_device=None):
    good_devices = int(num_devices * g)
    devices = [GoodWorker(beta)] * good_devices + [bad_worker] * (num_devices - good_devices)

    #aggregators = [GoodMeanAggregator(devices), MeanAggregator(), MedianAggregator(), TrimmedMean(), CCAggregator()]
    aggregators = [CCAggregator(1, 1), CCAggregator(10, 1), CCAggregator(1, 10), CCAggregator(10, 10)]

    errors = []
    accuracy = []

    for aggregator in aggregators:
        _, error, acc = distributed_sgd(aggregator, devices, good_devices, prob_class_1_per_device)
        errors.append(error)
        accuracy.append(acc)

    if plot == 'accuracy':
        plot_accuracy(aggregators, accuracy, beta, str(bad_worker))
    else:
        plot_criterion(aggregators, errors, beta, str(bad_worker))
```

### Эксперимент 1

- 2/10 устройств Византийцы
- Византийцы отправляют граденты со случайным шумом

```python
run_experiment(num_devices=10, g=0.8, beta=0, bad_worker=RandomNoiseAttack(noise_level=10), plot="criterion")
```

```python
run_experiment(num_devices=10, g=0.8, beta=0, bad_worker=RandomNoiseAttack(noise_level=10), plot="accuracy")
```
